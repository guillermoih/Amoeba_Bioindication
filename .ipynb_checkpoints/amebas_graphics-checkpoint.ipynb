# Imports

```python
%matplotlib inline

import umap

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.naive_bayes import CategoricalNB

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
random_s = 123
```

# Data loading

```python
df = pd.read_csv("/workspace/ameba/amebas_clasi.csv", delimiter=";")

transformed_df = df.pivot(index='muestra', columns='organismos', values=['porcentage','Alterado', 'ASV', 'otu'])

transformed_df["Alterado_new"] = transformed_df["Alterado"]["Arcellinida"]
transformed_df["ASV_new"] = transformed_df["ASV"]["Arcellinida"]
transformed_df["otu_new"] = transformed_df["otu"]["Arcellinida"]

transformed_df.drop("Alterado", axis=1, inplace=True)
transformed_df.drop("ASV", axis=1, inplace=True)
transformed_df.drop("otu", axis=1, inplace=True)

df_data = transformed_df[~transformed_df["Alterado_new"].isna()]
df_test = transformed_df[transformed_df["Alterado_new"].isna()]
```

```python
X = df_data["porcentage"]["Arcellinida"].to_numpy()
X_1 = np.transpose(np.vstack((X, df_data["porcentage"]["Diatomea"].to_numpy())))
#X = np.vstack((np.transpose(X),df_data["ASV_new"]))
#X = np.vstack((X, df_data["otu_new"]))

#X = np.transpose(X)

y_1 = df_data[["Alterado_new"]].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size=0.2,
                                                    random_state=random_s, stratify=y_1)
```

```python
X_train_0 = X_train[np.where(y_train==0)[0]]
X_train_1 = X_train[np.where(y_train==1)[0]]
X_test_0 = X_test[np.where(y_test==0)[0]]
X_test_1 = X_test[np.where(y_test==1)[0]]
```

# Show data

```python
fig = plt.figure(figsize = (10, 7))
 
# Creating plot
scatter = plt.scatter(X_1[:,0], X_1[:,1], c=y_1, cmap='cool')
fig.colorbar(scatter)

plt.show()
```

# Models

```python
names = [
    "Random Forest",
    "Decision Tree",
    "Adaboost",
    "SVM lineal",
    "SVM polinomic",
    "SVM rbf",
    "kNN"
]

classifiers = [
    RandomForestClassifier(random_state=random_s),
    DecisionTreeClassifier(random_state=random_s),
    AdaBoostClassifier(random_state=random_s),
    svm.SVC(kernel='linear', random_state=random_s),
    svm.SVC(kernel='poly', random_state=random_s),
    svm.SVC(kernel='rbf', random_state=random_s),
    KNeighborsClassifier()
]
```

# Colors

```python
cm = plt.cm.RdBu
cm_points = ListedColormap(["#404040"])
cm_guess = ListedColormap(["#1100ff", "#B30C0C"])

point_size = 500

marker_1 = '.'
marker_2 = '+'
```

```python
# Code source: Gaël Varoquaux
#              Andreas Müller
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

x_min, x_max = X_train[:, 0].min(), X_train[:, 0].max()
y_min, y_max = X_train[:, 1].min(), X_train[:, 1].max()

fig, ax = plt.subplots(3, len(classifiers) + 1, figsize=(7*len(classifiers), 21), dpi= 250)

# Plot the training and testing points
ax[0][0].set_title("Input data", size=26)

ax[0][0].scatter(
    X_train_0[:, 0], X_train_0[:, 1], c=np.zeros(len(X_train_0)),
    cmap=cm_points, edgecolors="k", s=point_size, marker=marker_1)
ax[0][0].scatter(
    X_train_1[:, 0], X_train_1[:, 1], c=np.ones(len(X_train_1)),
    cmap=cm_points, edgecolors="k", s=point_size, marker=marker_2)

ax[1][0].set_title("Test data", size=26)
ax[1][0].scatter(
    X_test_0[:, 0], X_test_0[:, 1], c=np.zeros(len(X_test_0)),
    cmap=cm_guess, alpha=0.6, edgecolors="k", s=point_size, marker=marker_1)
ax[1][0].scatter(
    X_test_1[:, 0], X_test_1[:, 1], c=np.ones(len(X_test_1)),
    cmap=cm_guess, alpha=0.6, edgecolors="k", s=point_size, marker=marker_2)

ax[0][0].set_xlim(x_min, x_max)
ax[0][0].set_ylim(y_min, y_max)
ax[1][0].set_xlim(x_min, x_max)
ax[1][0].set_ylim(y_min, y_max)

sns.set(font_scale=1.4)

# iterate over classifiers
i=1
for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    DecisionBoundaryDisplay.from_estimator(clf, X_train, cmap=cm, alpha=0.8, ax=ax[0][i], eps=0.5)

    # Plot the training and testing points
    ax[0][i].scatter(
        X_train_0[:, 0], X_train_0[:, 1], c=np.zeros(len(X_train_0)),
        cmap=cm_points, edgecolors="k", s=point_size, marker=marker_1)
    
    ax[0][i].scatter(
        X_train_1[:, 0], X_train_1[:, 1], c=np.ones(len(X_train_1)),
        cmap=cm_points, edgecolors="k", s=point_size, marker=marker_2)

    ax[0][i].set_xlim(x_min, x_max)
    ax[0][i].set_ylim(y_min, y_max)
    
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    ax[0][i].set_title(name + '\n Test accuracy: ' + str(round(acc, 4)), size=26)

    pred_guess = [int(a) ^ int(b) for a, b in zip(y_test, y_pred)]
    pred_guess_0 = np.array(pred_guess)[np.where(y_test==0)[0]]
    pred_guess_1 = np.array(pred_guess)[np.where(y_test==1)[0]]
    
    ax[1][i].scatter(X_test_0[:, 0], X_test_0[:, 1], c=pred_guess_0,
                     cmap=cm_guess, edgecolors="k", alpha=0.6, s=point_size, marker=marker_1)
    ax[1][i].scatter(X_test_1[:, 0], X_test_1[:, 1], c=pred_guess_1,
                     cmap=cm_guess, edgecolors="k", alpha=0.6, s=point_size, marker=marker_2)

    ax[1][i].set_xlim(x_min, x_max)
    ax[1][i].set_ylim(y_min, y_max)
    
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True,
                cmap='Blues', fmt = 'd', ax=ax[2][i], cbar=False)

    ax[2][i].set_ylabel("Real label")
    ax[2][i].set_xlabel("Predicted label")
    
    i += 1

# Hide plot in row 2 column 0
ax[2,0].axis('off')

plt.tight_layout()
plt.show()
```
